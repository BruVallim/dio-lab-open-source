```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

imigrantes_canada = pd.read_csv("/content/canadian_immegration_data.csv")
```

```python
imigrantes_canada.head()
```

```python
imigrantes_canada.set_index('Country', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
brasil = imigrantes_canada.loc ["Brazil", anos]
```

```python
br_imigrantes = {'Ano' : brasil.index.tolist(), 'Imigrantes' : brasil.values.tolist()}
```

```python
df_imigrantes_brasil = pd.DataFrame(br_imigrantes)
```

```python
df_imigrantes_brasil
```

```python
plt.plot(df_imigrantes_brasil['Ano'], df_imigrantes_brasil['Imigrantes'])
plt.title('Imigracao do Brasil para o Canada')
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.figure(figsize=(9,6))
plt.show()
```

```python
imigrantes_canada = pd.read_csv("/content/canadian_immegration_data.csv")
imigrantes_canada.set_index('Country', inplace=True)
anos = list(map(str, range(1980, 2014)))

brasil = imigrantes_canada.loc["Brazil", anos]
argentina = imigrantes_canada.loc["Argentina", anos]

plt.figure(figsize=(12, 6))
plt.plot(anos, brasil.values, label='Brasil')
plt.plot(anos, argentina.values, label='Argentina')

plt.title('ImigraÃ§Ã£o para o CanadÃ¡: Brasil vs. Argentina (1980-2013)')
plt.xlabel('Ano')
plt.ylabel('NÃºmero de Imigrantes')
plt.xticks(anos[::5]) # mostrar rotulos a cada 5 anos
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4)) #figura
ax.plot(df_imigrantes_brasil['Ano'], df_imigrantes_brasil['Imigrantes'], label='Brasil')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title ("Imigracao do Brasil para o Canada \n 1980 a 2013")
ax.set_xlabel("Ano")
ax.set_ylabel("Imigrantes")
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize= (15,5)) # duas figuras

axs[0].plot(df_imigrantes_brasil['Ano'], df_imigrantes_brasil['Imigrantes'])
axs[0].set_title('Imigracao do Brasil para o Canada \n 1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(df_imigrantes_brasil['Imigrantes'])
axs[1].set_title('imigracao para o canada \n 1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Imigrantes')
axs[1].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(2,2 , figsize=(15,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigracao dos quatro maiores paises da america do sul para o canada \n 1980 a 2013')

axs[0,0].plot(imigrantes_canada.loc['Brazil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(imigrantes_canada.loc['Colombia', anos])
axs[0,1].set_title('Colombia')

axs[1,0].plot(imigrantes_canada.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(imigrantes_canada.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Imigrantes')
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.grid()

ymin= 0
ymax= 7000
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)
```
